```python
import pandas as pd
import glob
import os

Excel_path=r'C:\Users\Tim\Desktop\Files_Python\FinancialD'
company_data={}
excel_files = [f for f in os.listdir(Excel_path) if f.endswith('.xlsx')]
excel_files
   
```

```python
file_data = {}

# Loop through each file in the specified folder
for file_name in excel_files:
    # Check if the file is an Excel file
    if file_name.endswith('.xlsx'):
        # Generate the full file path
        file_path = os.path.join(Excel_path, file_name)

    
        # Read the first two sheets of the Excel file into a dictionary of DataFrames
        df_sheets = pd.read_excel(file_path, sheet_name=[0, 1])  # Reads the first two sheets

        # Add the dictionary of sheets to `file_data` with the file name as the key
        file_data[file_name] = {
                "Sheet1": df_sheets[0],  # First sheet data
                "Sheet2": df_sheets[1]   # Second sheet data
            }
```

```python
file_data.items()
```

```python
dataframes = []

# Loop through each company in the file_data
for company_name, sheets in file_data.items():  # Here, company_name is the key
    # Access "Sheet1"
    if "Sheet1" in sheets:
        sheet1_data = sheets["Sheet1"]
        
        # Create a DataFrame for "Sheet1"
        df = pd.DataFrame(sheet1_data)
        
        # Add a new column for the company name using the key
        df['Company'] = company_name  # Use the key (company name) as the value for the new column
        
        # Append the DataFrame to the list
        dataframes.append(df)

# Concatenate all DataFrames vertically
final_dataframe = pd.concat(dataframes, ignore_index=True)
```

```python
final_dataframe
```

```python
output_file = "IncomeStatement.xlsx"  # Specify the name of the output file
final_dataframe.to_excel(output_file, index=False)  # Export without the index
```

```python
filtered_incomestate=income_state_df[income_state_df['Financial Indicators'].isin(['Net Income','Operating Revenue'])]
df=filtered_incomestate
df
```

```python
df['Net Income']=df.apply(lambda row: row['Value'] if row['Financial Indicators']== 'Net Income' else 0, axis=1)
df['Operating Revenue']=df.apply(lambda row: row['Value'] if row['Financial Indicators']== 'Operating Revenue' else 0, axis=1)
```

```python
df=df.drop(columns=['Financial Indicators'])
```

```python
df=df.drop(columns=['Value'])
```

```python
output_file = "Incomestate_filter.xlsx"  # Specify the name of the output file
df.to_excel(output_file, index=False)  # Export without the index
```

```python
modified_rows = []

# Loop through the DataFrame
for i in range(len(df)):
    if i % 2 == 0:  # Keep the odd rows (0-indexed means 0, 2, 4... are even)
        modified_rows.append(df.iloc[i])  # Add the odd row to the new list
        if i + 1 < len(df):  # Check if the next row exists
            modified_rows[-1]['Operating Revenue'] = df.iloc[i + 1]['Operating Revenue']  # Replace 0 with the even row's value

# Create a new DataFrame from the modified rows
modified_df = pd.DataFrame(modified_rows)
```
